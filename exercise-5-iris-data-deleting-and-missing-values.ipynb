### Step 3. Assign it to a variable called iris

```python
iris=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",sep=',')
iris.head()
```

### Step 4. Create columns for the dataset

```python
# 1. sepal_length (in cm)
# 2. sepal_width (in cm)
# 3. petal_length (in cm)
# 4. petal_width (in cm)
# 5. class
```

```python
list1=['sepal_length (in cm)','sepal_width (in cm)','petal_length (in cm)','petal_width (in cm)','class']
list1
```

```python
iris.columns=list1
iris.head()
```

### Step 5.  Is there any missing value in the dataframe?

```python
iris.isnull().sum()
```

### Step 6.  Lets set the values of the rows 10 to 29 of the column 'petal_length' to NaN

```python
#1st Method
iris['petal_length (in cm)'][10:30] = np.nan
```

```python
iris.iloc[10:30]
```

```python
#2nd Method
iris.iloc[10:30,2]=np.nan
iris.iloc[10:30,0:5]
```

```python
iris.isna().sum()
```

### Step 7. Good, now lets substitute the NaN values to 1.0

```python
iris.iloc[10:30,2].fillna(1.0,inplace = True)
iris.iloc[10:30,0:5]
```

### Step 8. Now let's delete the column class

```python
iris.drop(labels='class', axis=1)
```

### Step 9.  Set the first 3 rows as NaN

```python
iris.iloc[0:3,:]=np.nan
iris.iloc[0:3,:]
```

```python
iris.isna().sum()
```

### Step 10.  Delete the rows that have NaN

```python
iris=iris.dropna(axis = 0, how = "any")
iris.head(10)
```

### Step 11. Reset the index so it begins with 0 again

```python
iris = iris.reset_index(drop = True)
iris.head(100)
```
